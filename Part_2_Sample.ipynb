# We will now perform interpolation, estimating the missing values in the column based on their neighbouring data points


```python
df['ActualElapsedTime'] = df['ActualElapsedTime'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df['CRSElapsedTime'] = df['CRSElapsedTime'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df['AirTime'] = df['AirTime'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df['ArrTime'] = df['ArrTime'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df['ArrDelay'] = df['ArrDelay'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df['CarrierDelay'] = df['CarrierDelay'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df['WeatherDelay'] = df['WeatherDelay'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df['NASDelay'] = df['NASDelay'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df['LateAircraftDelay'] = df['LateAircraftDelay'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df['SecurityDelay'] = df['SecurityDelay'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df['TaxiIn'] = df['TaxiIn'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df['TaxiOut'] = df['TaxiOut'].interpolate(method='linear', limit_direction='both', axis=0) #Interpolation
df.isna().sum()
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['DepTime', 'TailNum'])
df
```

# We changed the datatype of these features to the best suited datatype as per the data that they contain

```python
# Converting Categorical Variables to object type
df['Year'] = df['Year'].astype(object)
df['Month'] = df['Month'].astype(object)
df['DayofMonth'] = df['DayofMonth'].astype(object)
df['DayOfWeek'] = df['DayOfWeek'].astype(object)

df['UniqueCarrier'] = df['UniqueCarrier'].astype(object)
df['FlightNum'] = df['FlightNum'].astype(object)
df['TailNum'] = df['TailNum'].astype(object)

df['Origin'] = df['Origin'].astype(object)
df['Dest'] = df['Dest'].astype(object)

df['Cancelled'] = df['Cancelled'].astype(object)
df['CancellationCode'] = df['CancellationCode'].astype(object)
df['Diverted'] = df['Diverted'].astype(object)
```

# Here I am dropping Cancellation code as I dont think this would be a significant feature what will help us in our prediction

```python
df.drop('CancellationCode', axis=1, inplace=True)
```

```python
df['Delay'] = ''
```

```python
df['Delay'] = np.where((df['ArrDelay'] <= 0) & (df['DepDelay'] <= 0), 'N', 'Y')
```
